```python
def rank(r):
    return r.split("\\")[1].split("-")[0]
```

```python
import pandas as pd
import glob
csv_f = pd.Series(glob.glob("./IG/*.csv"))
df = pd.DataFrame({"rank":csv_f.apply(rank), "path":csv_f})
df1 = df[[int(r) // 10 == 0 for r in df["rank"]]]
df2 = df.drop(df1.index)
df = pd.concat([df1, df2]).reset_index(drop=True)
df
```

```python
import os

name_list = []
image_num_list = []
video_num_list = []

for f in df["path"]:
    df3 = pd.read_csv(f)
#     name = f.split("\\")[1].split("-")[1].split(".")[0]
    name = os.path.splitext(f)[0].split("\\")[1].split("-")[1]
    name_list.append(name)
    bool_filiter = df3["類別"].str.contains("Image") | df3["類別"].str.contains("Sidecar")
    image_num = sum([len(u.split(",")) for u in df3[bool_filiter]["影像連結"]])
    image_num_list.append(image_num)
    video_num = sum([len(u.split(",")) for u in df3[[not b for b in bool_filiter]]["影像連結"]])
    video_num_list.append(video_num)
df["name"] = name_list
df["image_num"] = image_num_list
df["video_num"] = video_num_list
df
```

```python
df_all = pd.read_csv("ALL.csv")
df_class = df_all[["名次", "分類"]]
df["分類"] = df_class[[str(r) in list(df["rank"]) for r in df_class["名次"]]].reset_index(drop=True)["分類"]
df = df.drop(["path"], axis=1)
df = df[["分類", "rank", "name", "image_num", "video_num"]]
df
# df.to_csv("IG_num.csv", index=False)
```

```python
classes = list(set(df["分類"]))
num_list = []
image_num_list = []
video_num_list = []
for c in classes:
    df_sub = df[[str(r) in c for r in df["分類"]]].reset_index(drop=True)
    num_list.append(df_sub.shape[0])
    image_num_list.append(df_sub["image_num"].sum())
    video_num_list.append(df_sub["video_num"].sum())
df_final = pd.DataFrame([classes, num_list, image_num_list, video_num_list]).T
df_final.columns = ["分類", "人數", "圖片數量", "影片數量"]
df_final = df_final.sort_values(by=["圖片數量"], ascending=False).reset_index(drop=True)
df_final

# df_final.to_csv("IG_num_class.csv")
```

```python
classes = list(df_final["分類"])
df_sub_all = pd.DataFrame()
for c in classes:
    df_sub = df[[str(r) in c for r in df["分類"]]].reset_index(drop=True)
    for c in  df_sub.columns[3:5]:
        df_sub[c+"_percent"] = round(df_sub[c] / df_sub[c].sum() * 100, 2)
    df_sub_all = pd.concat([df_sub_all, df_sub])
pd.set_option('display.max_rows', 100)
df_sub_all = df_sub_all[["分類", "rank", "name", "image_num", "video_num", "image_num_percent", "video_num_percent"]].reset_index(drop=True)
df_sub_all

# df_sub_all.to_csv("IG_num_percent.csv")
```
